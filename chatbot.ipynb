```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
embeddings = model.encode(sentences)

print(embeddings)
```

```python
df = pd.read_csv('C:/Users/lg/Downloads/전국관광지정보표준데이터.csv' , encoding='cp949')

df.head()
```

```python
df = df.drop(columns=['소재지지번주소','지원시설정보'])

df.head()
```

```python
df = df[~df['소재지도로명주소'].isna()]
        
df.head()
```

```python
df.loc[0, '관광지소개']
```

```python
df['embedding'] = pd.Series([[]] * len(df))

df['combined'] = df['관광지소개'] + ' ' + df['관광지명'] + ' ' + df['제공기관명']

df['embedding'] = df['combined'].map(lambda x: list(model.encode(x)))

df.head()
```

```python
df.to_csv('tour_dataset.csv', index=False)
```

```python
text = '광주 산에 놀러 가고 싶어'

embedding = model.encode(text)
```

```python
df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())

df.head()
```

```python
top_5 = df.nlargest(5, 'distance')

for index, row in top_5.iterrows():
    print('관광지명:', row['관광지명'])
    print('관광지 소개:', row['관광지소개'])
    print('주소:', row['소재지도로명주소'])
    print('편의시설:', row['공공편익시설정보'])
    print('유사도:', row['distance'])
    print('--------------------')
```
